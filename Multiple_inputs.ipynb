# Multiple Inputs Graph

A LangGraph agent that takes a list of integers, a name, and an operation (`+` or `*`), then computes the result in a single node.

```python
import os
from typing import TypedDict
from dotenv import load_dotenv

from langgraph.graph import StateGraph, END

load_dotenv()
```

```python
class State(TypedDict):
    name: str
    numbers: list[int]
    operation: str
    result: float
```

```python
def compute(state: State) -> State:
    nums = state["numbers"]
    op = state["operation"]

    if op == "+":
        computed = sum(nums)
    elif op == "*":
        computed = 1
        for n in nums:
            computed *= n
    else:
        raise ValueError(f"Unknown operation: {op}")

    return {"result": computed}
```

```python
graph = StateGraph(State)
graph.add_node("compute", compute)
graph.set_entry_point("compute")
graph.add_edge("compute", END)

app = graph.compile()
```

```python
print("Addition:")
r = app.invoke({"name": "Abelle", "numbers": [3, 1, 4, 1, 5], "operation": "+"})
print(f"{r['name']}, the sum is: {r['result']}")

print()

print("Multiplication:")
r = app.invoke({"name": "Abelle", "numbers": [3, 1, 4], "operation": "*"})
print(f"{r['name']}, the product is: {r['result']}")
```
