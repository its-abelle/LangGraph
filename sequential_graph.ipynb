# Sequential Graph

A 3-node LangGraph pipeline that accepts a name, age, and list of skills, then processes them sequentially and combines the output.

```python
import os
from typing import TypedDict
from dotenv import load_dotenv

from langgraph.graph import StateGraph, END

load_dotenv()
```

```python
class State(TypedDict):
    name: str
    age: int
    skills: list[str]
    greeting: str
    age_description: str
    skills_list: str
    result: str
```

```python
def add_greeting(state: State) -> State:
    return {"greeting": f"Hello, {state['name']}! Welcome aboard."}


def describe_age(state: State) -> State:
    return {"age_description": f"You are {state['age']} years old — a great age to be learning."}


def list_skills(state: State) -> State:
    formatted = "\n".join(f"  • {skill}" for skill in state["skills"])
    skills_text = f"Your skills:\n{formatted}"
    result = f"{state['greeting']}\n{state['age_description']}\n{skills_text}"
    return {"skills_list": skills_text, "result": result}
```

```python
graph = StateGraph(State)
graph.add_node("add_greeting", add_greeting)
graph.add_node("describe_age", describe_age)
graph.add_node("list_skills", list_skills)

graph.set_entry_point("add_greeting")
graph.add_edge("add_greeting", "describe_age")
graph.add_edge("describe_age", "list_skills")
graph.add_edge("list_skills", END)

app = graph.compile()
```

```python
result = app.invoke({
    "name": "Abelle",
    "age": 25,
    "skills": ["Python", "LangGraph", "Jupyter", "Git", "Solidity", "Foundry"],
})
print(result["result"])
```
