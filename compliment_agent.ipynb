```python
import os
from typing import TypedDict
from dotenv import load_dotenv

from langgraph.graph import StateGraph, END
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
print("API key loaded:", "yes" if os.environ.get("GOOGLE_API_KEY") else "no")
```

```python
class State(TypedDict):
    name: str
    compliment: str
    result: str
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=os.environ["GOOGLE_API_KEY"],
)
```

```python
def generate_compliment(state: State) -> State:
    prompt = (
        f"Generate a short, encouraging one-sentence compliment for {state['name']} "
        f"who is learning LangGraph. Make it motivating and personal."
    )
    response = model.invoke(prompt)
    compliment = response.content
    return {
        "compliment": compliment,
        "result": f"{state['name']} - {compliment}",
    }
```

```python
graph = StateGraph(State)
graph.add_node("generate_compliment", generate_compliment)
graph.set_entry_point("generate_compliment")
graph.add_edge("generate_compliment", END)

app = graph.compile()
```

```python
result = app.invoke({"name": "Abelle"})
print(result["result"])
```
